```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import yfinance as yf
gold = yf.download('GC=F', end='2024-08-27')['Close']
sp500 = yf.download('SPY', end='2024-08-27')['Close']
reits = yf.download('VNQ', end='2024-08-27')['Close']
bonds = yf.download('ZB=F', end='2024-08-27')['Close']
int_stocks = yf.download('EFA', end='2024-08-27')['Close']
oil = yf.download('USO', end='2024-08-27')['Close']
df = pd.concat([gold,sp500,reits,bonds,int_stocks,oil], ignore_index=True, axis=1, join='outer')
df.reset_index(inplace=True)
df.rename(columns={'Date':'DATE'},inplace=True)
df.rename(columns={0:'Gold', 1:'S&P 500', 2:'REITs', 3:'Bonds', 4:'International Stocks',5:'Crude Oil'},inplace=True)
# Loop through each asset column
for column in df.iloc[:,1:].columns:
    # Calculate the percentage change (returns) for each asset
    df[f'{column}_returns'] = df[column].pct_change()
    df[f'{column}_roll'] = df[f'{column}_returns'].rolling(window=30).mean()
    df[f'{column}_std'] = df[f'{column}_returns'].rolling(window=30).std()
df.dropna()
```

```python
df2 = df[(df[f'{column}_returns']<(df[f'{column}_roll']-2*df[f'{column}_std']))]
df2
```

```python
df.dropna(inplace=True)
df.to_csv('Asset Class.csv')
df2.to_csv('Tail.csv')
```

```python
df = pd.read_csv('Asset Class.csv')
df2 = pd.read_csv('Tail.csv')
```

```python
pct = [0,5,10,15]
result1 = pd.DataFrame(columns=['Annualized Excess Returns','Annualized Standard Deviation', 'Sharpe Ratio','Beta','Treynor Measure','VaR','CVaR'])
m = df['S&P 500_returns']-df['Bonds_returns']
m_ret = np.mean(m)
for i in pct:
    ret = df['Gold_returns']*(i*0.01)+df['S&P 500_returns']*(1-i*0.01)
    ex_ret = ret - df['Bonds_returns']
    var = np.percentile(ex_ret, 5)
    s_ret = np.sort(ex_ret) 
    n = int(0.05 * len(s_ret))
    cvar = (np.mean(s_ret[:n]))
    result1.loc[f'{i}% Gold']=[np.mean(ex_ret)*252,np.std(ex_ret)*252**0.5, (np.mean(ex_ret)/np.std(ex_ret))*252**0.5, np.cov(m,ex_ret)[0, 1]/np.var(m),
                                  np.mean(ex_ret)*252/(np.cov(m,ex_ret)[0, 1]/np.var(m)), var, cvar] 
result1
```

```python
from scipy.optimize import minimize
def portfolio_stats(weights, returns):
    portfolio_return = np.sum(returns.mean() * weights)
    portfolio_vol = np.sqrt(np.dot(weights.T, np.dot(returns.cov(), weights)))
    return np.array([portfolio_return, portfolio_vol, portfolio_return/portfolio_vol])

def neg_sharpe_ratio(weights, returns):
    return -portfolio_stats(weights, returns)[2]

def optimize_portfolio(returns):
    num_assets = returns.shape[1]
    args = (returns,)
    constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
    bounds = tuple((0, 1) for asset in range(num_assets))
    
    result = minimize(neg_sharpe_ratio, num_assets*[1./num_assets], args=args,
                      method='SLSQP', bounds=bounds, constraints=constraints)
    
    return result
```

```python
returns = df[['S&P 500_returns', 'REITs_returns', 
              'Bonds_returns', 'International Stocks_returns', 'Crude Oil_returns']]
p = optimize_portfolio(returns)
p.x
```
